```python
df = pandas.read_csv('../ex2/proc.txt', sep=';').drop(columns=['sum', 'pi']).rename(columns={'n_proc': 'n_processors'})
# df = df.pivot(index="Message size", columns="Communication type", values='Throughput')
# df.plot(title="Two nodes throughput")
df
```

```python
df = df.pivot(index="n_processors", columns="n_points", values='time')
df
```

```python
df.plot(figsize=(15,10), title="Time", style='.-')
```

```python
df.plot(figsize=(15,10), logy=True, title="Time log", style='.-')
```

```python
sdf = df.diff() * -1
```

```python
sdf.plot(figsize=(15,10))
```

```python
(df.pct_change() * -1).plot(figsize=(15,10))
```

```python
df.div(df.shift(-1)).fillna(df).plot()
```

```python
sdf = df.apply(lambda x: 1/x).mul(df.iloc[(0)])
sdf
```
