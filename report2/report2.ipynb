## Part 3

```python
rdf = pd.read_csv('../lab3/time.csv', sep=';')
df = rdf.groupby('nt').mean()
df
```

```python
df.plot(logy=False, legend=True,figsize=(15,10), title="Time (s): per section", style='.-', xticks=range(1,5))
```

```python
df.plot(logy=True, legend=True,figsize=(15,10), title="Time (log(s)): per section", style='.-', xticks=range(1,5))
```

```python
sdf = df.apply(lambda x: 1/x).mul(df.iloc[(0)])
sdf
```

```python
sdf.plot(logy=False, legend=True,figsize=(15,10), title="Speedup: per section", style='.-', xticks=range(1,5))
```
